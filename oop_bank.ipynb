```python
class Bank:

    def __init__(self, name,accounts={}):
        self.name = name
        self.accounts = accounts


    def add_account(self):
        username = input("Please enter your name: ")
        try:
            balance = float(input("Please enter your balance: "))
            self.accounts[username] = balance
            print(f"Account for {username} added successfully")
        except ValueError:
            print("Invalid balance. Please enter a number")

    def show_balance(self, name):
        if name in self.accounts:
            print(f"{name}'s balance: {self.accounts[name]}")
        else:
            print("Account not found")

    def deposit(self, name):
      if name not in self.accounts:
        print("Account not found.")
        return
      while True:
        try:
          amount = float(input("Please enter the amount to deposit: "))
          if amount <= 0:
            print("Invalid amount. Please enter a positive number")
            continue
        except ValueError:
          print("Invalid amount. Please enter a number")
          continue

        else:
          self.accounts[name]+=amount
          print(f"Deposit of {amount} successful")
          break

    def withdraw(self, name):
      if name not in self.accounts:
        print("Account not found.")
        return
      while True:
        try:
          amount = float(input("Please enter the amount to withdraw: "))
          if amount <= 0:
            print("Invalid amount. Please enter a positive number")
            continue
          elif amount > self.accounts[name]:
                print("Insufficient funds.")
          else:


            self.accounts[name] -= amount
            print(f"Withdrawal of {amount} successful. New balance: {self.accounts[name]}")
            break
        except ValueError:
          print("Invalid amount. Please enter a number")



```

```python
import random


bank_names = ["Future Bank", "SmartSaver", "PixelBank", "Quantum Credit", "NeoFinance"]
random_bank = random.choice(bank_names)
bank = Bank(random_bank)

while True:
    print(f"\n====== Welcome to {bank.name} ======")
    print("1. Add Account")
    print("2. Show Balance")
    print("3. Deposit")
    print("4. Withdraw")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        bank.add_account()
    elif choice == "2":
        name = input("Enter your name: ")
        bank.show_balance(name)
    elif choice == "3":
        name = input("Enter your name: ")
        bank.deposit(name)
    elif choice == "4":
        name = input("Enter your name: ")
        bank.withdraw(name)
    elif choice == "5":
        print("Thank you for banking with us Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
